```octave
// if and else

function [z] = h(x, y)
    if x > y & y <> 0 then
        z = x/y
    elseif x == y then
        z = 1
    else
        z = x*y
    end
endfunction

h(2, 1)
h(3, 3)
h(2, 3)

// 3D Plotting
x = linspace(-10, 10, 100);
y = linspace(-10, 10, 100);
z = feval(x, y, h)';
// surf(x, y, z)
```

Usaremos o Teorema de Bolzano para concluir que existe pelo menos uma solucao real para a equacao:
   
`e^x = x + 2`

```octave
deff('y=f(x)', 'y=exp(x)-x-2')
f(-2),f(0)
sign(f(-2)),sign(f(0))
```

Para garantir que o zero seja unico, a funcao f precisa satisfazer os seguintes criterios para um dado intervalo [a, b]:
- `f(a)*f(b) < 0`
- `f'(x) < 0 || f'(x) > 0`

```octave
// Podemos observar o comportamento de f no dado intervalo
x = linspace(-2, 0, 50);
plot(x, f(x));xgrid
deff('y=fl(x)', 'y=exp(x)-1')
plot(x, fl(x), 'r');xgrid
```
