```octave
funcprot(0);    // Permite redefinir funções

// Definindo funções
function [y] = f(x)    // Melhor para funções com várias variáveis
    y = x^2
endfunction

deff('y = g(x)', 'y = sin(x)')    // Melhor para funções com poucas variáveis

// Gráficos de funções
//  linspace(inicio, fim, pontos)
x = linspace(0, 1, 100);    // Cria o eixo-x
plot(x, f(x), "b", x, g(x), "r"); xgrid
```

```octave
// Laço for
x = 0;
for i = 2:2:10
    x = x + i;
end
x

// Laço while
s = 0; i = 1;
while ( i <= 10 )
    s = s + i;
    i = i +1;
end
s
i
```

```octave
// if and else

function [z] = h(x, y)
    if x > y & y <> 0 then
        z = x/y
    elseif x == y then
        z = 1
    else
        z = x*y
    end
endfunction

h(2, 1)
h(3, 3)
h(2, 3)

// 3D Plotting
x = linspace(-10, 10, 100);
y = linspace(-10, 10, 100);
z = feval(x, y, h)';
// surf(x, y, z)
```
